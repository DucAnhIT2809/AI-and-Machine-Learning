```python
from ast import literal_eval
def count_item(items):
  count = np.zeros(50,dtype =np.int8)
  for timestep in range(50):
    for item in items:
      if item[0] <= timestep +1:
        count[timestep] +=1
  return count     
df_notna['bDragons'] = df_notna['bDragons'].apply(literal_eval)
df_notna['rDragons'] = df_notna['rDragons'].apply(literal_eval)
df_notna['bDragons'] = df_notna['bDragons'].apply(count_item)
df_notna['rDragons'] = df_notna['rDragons'].apply(count_item)
df_notna['dragondiff'] = df_notna['bDragons']-df_notna['rDragons']
```

```python
df_notna['dragondiff'].head(10)
```

```python
df_notna['bKills'] = df_notna['bKills'].apply(literal_eval)
df_notna['rKills'] = df_notna['rKills'].apply(literal_eval)
df_notna['bKills'] = df_notna['bKills'].apply(count_item)
df_notna['rKills'] = df_notna['rKills'].apply(count_item)
df_notna['killdiff'] = df_notna['bKills']-df_notna['rKills']
df_notna['killdiff']
```

```python
df_notna['bTowers'] = df_notna['bTowers'].apply(literal_eval)
df_notna['rTowers'] = df_notna['rTowers'].apply(literal_eval)
df_notna['bTowers'] = df_notna['bTowers'].apply(count_item)
df_notna['rTowers'] = df_notna['rTowers'].apply(count_item)
df_notna['Towersdiff'] = df_notna['bTowers']-df_notna['rTowers']
```

```python
df_notna['bInhibs'] = df_notna['bInhibs'].apply(literal_eval)
df_notna['rInhibs'] = df_notna['rInhibs'].apply(literal_eval)
df_notna['bInhibs'] = df_notna['bInhibs'].apply(count_item)
df_notna['rInhibs'] = df_notna['rInhibs'].apply(count_item)
df_notna['Inhibsdiff'] = df_notna['bInhibs']-df_notna['rInhibs']
```

```python
df_notna['bBarons'] = df_notna['bBarons'].apply(literal_eval)
df_notna['rBarons'] = df_notna['rBarons'].apply(literal_eval)
df_notna['bBarons'] = df_notna['bBarons'].apply(count_item)
df_notna['rBarons'] = df_notna['rBarons'].apply(count_item)
df_notna['Baronsdiff'] = df_notna['bBarons']-df_notna['rBarons']
```

```python
df_notna['bHeralds'] = df_notna['bHeralds'].apply(literal_eval)
df_notna['rHeralds'] = df_notna['rHeralds'].apply(literal_eval)
df_notna['bHeralds'] = df_notna['bHeralds'].apply(count_item)
df_notna['rHeralds'] = df_notna['rHeralds'].apply(count_item)
df_notna['Heraldsdiff'] = df_notna['bHeralds']-df_notna['rHeralds']
```
